```python
#nambahin fitur Cek Stok dan Pembayaran

# Mendefinisikan harga buah
price_apel = 10000
price_jeruk = 15000
price_anggur = 20000

#stok buah
S_Apel = 5
S_Jeruk = 5
S_Anggur =5

#Menghitung stok
while True:
    n_apel = int(input('Masukkan jumlah apel: '))
    if  n_apel > S_Apel:
        print(f"stok apel sisa {S_Apel}, masukkan stok kembali")
    else: 
        print(f"stok diterima, lanjutkan input buah")
        break
    
while True:
    n_jeruk = int(input('Masukkan jumlah jeruk: '))
    if n_jeruk > S_Jeruk:
        print(f"stok jeruk sisa {S_Jeruk}, masukkan stok kembali")
    else: 
        print(f"stok diterima, lanjutkan input buah")
        break   

while True: 
    n_anggur = int(input('Masukkan jumlah anggur: '))
    if n_anggur > S_Jeruk:
        print(f"stok anggur sisa {S_Anggur}, masukkan stok kembali")
    else: 
        print(f"stok diterima, sedang menghitung")
        break


# Menghitung total belanjaan per buah
total_apel = n_apel * price_apel
total_jeruk = n_jeruk * price_jeruk
total_anggur = n_anggur * price_anggur

# Menghitung total belanjaan
total = total_apel + total_jeruk + total_anggur
      
#print total belanjaan
print(f' {n_apel} Apel X {price_apel} = {total_apel}') 
print(f' {n_jeruk} Jeruk X {price_jeruk} = {total_jeruk}') 
print(f' {n_anggur} Anggur X {price_anggur} = {total_anggur}') 
print(f' {total}')
      
while True:
    bayar = int (input (f' masukkan pembayaran anda: '))
    if bayar >= total:
        print(f' terima kasih')
        print(f' Uang kembalian anda = Rp{bayar - total}')
        break
    else :
        print (f' Uang anda kurang Rp{total - bayar}')

# stok sisa
S_Apel -= n_apel 
S_Jeruk -= n_jeruk 
S_Anggur -= n_anggur 

print (f' sisa stok saat ini = {S_Apel} Apel, {S_Jeruk} Jeruk, {S_Anggur} Anggur ')
```
